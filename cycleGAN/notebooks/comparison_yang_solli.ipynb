#open the files and get a histogram for both

```python
dfYang = np.loadtxt("paletteScoresKulerYang.csv")
dfSolli = np.loadtxt("paletteScoresKulerSolli.csv")
```

```python
idxsort = np.argsort(dfYang)
dfSolli = dfSolli[idxsort]
dfYang = dfYang[idxsort]
```

```python
plt.figure(figsize=(10,11))
plt.suptitle("Comparaison des deux scores d'harmonie sur des palettes",fontsize=30)

plt.subplot(2,1,1)
plt.title("Méthode adaptée à partir de Yang et al.",fontsize=20)
plt.xlabel("index de la palette")
plt.ylabel("score d'harmonie")
plt.plot(dfYang)

plt.subplot(2,1,2)
plt.title("Méthode adaptée à partir de Solli et Renz",fontsize=20)
plt.xlabel("index de la palette")
plt.ylabel("score d'harmonie")
plt.plot(dfSolli)
```

#Compute pearson coefficient

```python
twoArrays=np.concatenate([dfYang.reshape((len(dfYang),1)),dfSolli.reshape((len(dfSolli),1))],axis=1)
np.corrcoef(twoArrays.transpose())
```
